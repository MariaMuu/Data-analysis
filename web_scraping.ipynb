```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.scrapethissite.com/pages/forms'
```

```python
page = requests.get(url)
```

we will now take a python "snapshot" of the html website (inspect)

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
soup.find_all('p') #we get the text of the page
```

```python
soup.find_all('p', class_ = 'lead') #we get the specific class (text).
```

```python
soup.find('p', class_ = 'lead').text # we get only the text without programming stuff
```

```python
soup.find('p', class_ = 'lead').text.strip() #looks better now
```

```python
soup.find_all('table')
```

```python
table = soup.find('table')
print(table)
```

```python
soup.find('table').text.strip()
```

```python
soup.find_all('td') #we get the rows of the table
```

```python
soup.find_all('th') #we get the titles (columns of the table)
```

```python
columns = soup.find_all('th') #we get the columns
print(columns)
```

```python
soup.find('th').text.strip()
```

```python
columns_table = [column.text.strip() for column in columns]
print(columns_table)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = columns_table)
df
```

```python
column_data = table.find_all('tr') #we check how rows of the table are displayed, commas, etc.
```

# We will now create the table of the website

```python
for row in column_data[1:]: #we're looping through the 'tr' data
  row_data = row.find_all('td') #we're looking for our td tags (individual data)
  individual_row_data = [data.text.strip() for data in row_data] #we kind of clean the data and put it in a list
  #print(individual_row_data)

  length =  len(df)
  df.loc[length] = individual_row_data #we're appending each row of the information into the next position of the table
```
